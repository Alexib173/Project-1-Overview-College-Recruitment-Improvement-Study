## I use some AI to rename some of the columns
### prompt : I have a survey dataset with many similar questions like ‘How did you hear about County College of Morris?’ repeated for each option (e.g. [Website], [Social Media], etc.).

### I want to combine each group of related questions into one summary column (like Heard_From), keep a few key columns such as course, age, and gender, and remove all the detailed ones.

### Can you help me write clean Python code using pandas to do that automatically?”

```python
#Define columns group and separe the one that I want to group or join and the one that I want to keep and rename 
groups = {
    "How did you hear about County College of Morris": "Heard_From",
    "impact your decision to attend County College of Morris": "Impact_Reasons",
    "Prior to applying to college": "Prior_Activities",
    "activities or experience impact your decision to enroll": "Computing_Influences"
}

rename_keep = {
    "Timestamp": "timestamp",
    "Which course are you enrolled in?": "course_enrolled",
    "What degree program are you currently enrolled in?": "degree_program",
    "Was a computing major/certificate your first choice, or did you change majors from a different CCM program? If you changed majors, indicate what your first major was.": "first_choice_or_change",
    "On a scale of 1 to 5, with 1 being not at all interested and 5 being extremely interested, how interested are you in taking more computing classes?": "interest_more_classes_scale",
    "If you are interested in taking more computing classes, please explain why.": "interest_explanation",
    "Please explain your answer to the question above.  Why or why not would you be interested in taking another computing class?":"interest_explanation",
    "Gender": "gender",
    "Race/ethnicity": "ethnicity",
    "Age": "age"
}
```

```python
#taking the information of the brackets to separe 
def bracket_label(text):
    m = re.search(r"\[(.*?)\]", text)
    return m.group(1).strip() if m else text.strip()
```

### Here is where I got the most help from the AI to create an automated system to separe the columns and put it in rows 

```python


def combine_columns(prefix):
    def normalize(s):
        return re.sub(r"[^a-z0-9]", "", s.lower())

    normalized_prefix = normalize(prefix)
    cols = [c for c in df.columns if normalized_prefix in normalize(c)]
    if not cols:
        return None

    def join_row(row):
        answers = []
        for c, v in row.items():
            val = str(v).strip().lower()
            label = bracket_label(c)

            # Group-specific logic
            if any(x in prefix.lower() for x in ["how did you hear", "did you receive information", "motivations"]):
                if val in ["yes", "y", "true", "1"]:
                    answers.append(label)
            elif "prior to applying" in prefix.lower():
                if val in ["yes", "not sure"]:
                    answers.append(f"{label} ({v})")
            else:
                if val not in ["", "no impact", "n/a", "dont recall", "na", "no"]:
                    answers.append(f"{label} ({v})")
        return ", ".join(answers)

    return df[cols].apply(join_row, axis=1)
```

```python
for prefix, new_col in groups.items():
    df[new_col] = combine_columns(prefix)
```

```python
#drop all old columns 
df = df.drop(columns=[c for c in df.columns if any(p in c for p in groups)], errors="ignore")
```

```python
#rename and reorder columns 
df = df.rename(columns={k: v for k, v in rename_keep.items() if k in df.columns})
df_cleaned = [v for v in rename_keep.values() if v in df.columns] + list(groups.values())
df_cleaned = df[df_cleaned]
```
